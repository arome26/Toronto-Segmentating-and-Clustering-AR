```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
ddfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
df.columns
```

```python
df[1] = df[1].str.replace("\n","")
df.head()
```

```python
df = df.groupby([0,1])[2].apply(", ".join).reset_index()
df.head()
```

```python
print("Shape: ", df.shape)
```

```python
df_geo_coor = pd.read_csv("http://cocl.us/Geospatial_data")
```

```python
df_geo_coor.head()
```

```python
df_toronto = pd.merge(df, df_geo_coor, how='left', left_on = 0, right_on = "Postal Code")
```

```python
df_toronto.drop("Postal Code", axis=1, inplace=True)
df_toronto.head()
```

```python
address = "Toronto, ON"

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto city are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
map_toronto
```

```python
df_toronto=df_toronto.dropna()
for lat, lng, borough, neighborhood in zip(
        df_toronto['Latitude'], 
        df_toronto['Longitude'], 
        df_toronto[2], 
        df_toronto[1]):
    label = '{}, {}'.format([2], [1])
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

map_toronto
```

```python
# "denc" = [D]owntown Toronto, [E]ast Toronto, [N]orth Toronto, [C]entral Toronto
df_toronto_denc = df_toronto[df_toronto[2].str.contains("Toronto")].reset_index(drop=True)
df_toronto_denc.head()
```

```python
map_toronto_denc = folium.Map(location=[latitude, longitude], zoom_start=12)
for lat, lng, borough, neighborhood in zip(
        df_toronto_denc['Latitude'], 
        df_toronto_denc['Longitude'], 
        df_toronto_denc[2], 
        df_toronto_denc[1]):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto_denc)  

map_toronto_denc
```
